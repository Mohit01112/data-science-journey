```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## lets create some datas points
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_classes=2,n_features=2,n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
## prediction
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(x_train,y_train)
```

```python


# 3. Now you can safely PREDICT
y_pred1 = rbf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(x_train,y_train)
## prediction
y_pred2=polynomial.predict(x_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
## prediction
y_pred3=sigmoid.predict(x_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
from sklearn.model_selection import GridSearchCV

## defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
           'gamma':[1,0.1,0.01,0.001,0.0001],
           'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(x_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
