```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
## quadratic equation used y=0.5x^2+1.5x+2+outliers
plt.scatter(x,y,color='g')
plt.xlabel('x dataset')
plt.ylabel('y dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
regression_1.fit(x_train,y_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Re-split clean data ensuring the standard variable order
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 2. Check the lengths to confirm they match perfectly
print(f"Verified Match -> X_train: {len(x_train)} rows | y_train: {len(y_train)} rows")

# 3. Fit your model using the verified matching variables
regression_1 = LinearRegression()
regression_1.fit(x_train, y_train)
print("Model trained successfully!")

```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression_1.predict(x_test))
print(score)
```

```python
## lets visualize
plt.plot(x_train,regression_1.predict(x_train),color='r')
plt.scatter(x_train,y_train)
plt.xlabel("x dataset")
plt.ylabel*("y dataset")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

def poly_regression(degree):
    # 1. Create dummy x values for smooth curve plotting
    x_new = np.linspace(-3, 3, 200).reshape(200, 1)

    # 2. Fixed: Used the dynamic input 'degree' argument here instead of hardcoded 2
    poly_features = PolynomialFeatures(degree=degree, include_bias=True)
    lin_reg = LinearRegression()
    
    # 3. Fixed: Changed pipeline variable name to 'pipeline_model' to avoid function conflict
    pipeline_model = Pipeline([
        ("poly_features", poly_features),
        ("lin_reg", lin_reg)
    ])
    
    # 4. Train the model and make predictions
    pipeline_model.fit(x_train, y_train)
    y_pred_new = pipeline_model.predict(x_new)
    
    ## Plotting components
    # 5. Fixed: Resolved the linewidth assignment typo and consolidated the labels
    plt.plot(x_new, y_pred_new, 'r', linewidth=2, label=f"Degree {degree} Predictions")

    plt.plot(x_train, y_train, "b.", label='Training points')
    plt.plot(x_test, y_test, "y.", label='Testing points')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
```

```python
poly_regression(10)
```
