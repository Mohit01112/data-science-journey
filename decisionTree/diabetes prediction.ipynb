```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset.DESCR) 
```

```python
dataset.feature_names
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df_diabetes.head()
```

```python
x=df_diabetes
```

```python
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
import seaborn as sns
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
## hyperparamter tunning
```

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error'],'splitter':['best','random'],'max_depth':[1,2,3,4,5,10,15,20,25],'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=params,cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
