```python
## train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
## create  column transfiomer with 3 types of transformers
cat_features=x.select_dtypes(include="object").columns
num_features=x.select_dtypes(include="object").columns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(drop='first')
```

```python
preprocessor=ColumnTransformer(
    [
        ("OneHotEncoder",oh_transformer,cat_features)
        ,("StandardScaler",numeric_transformer,num_features)
    ]
)
```

```python
preprocessor
```

```python
x_train=preprocessor.fit_transform(x_train)
```

```python
pd.DataFrame(x_train)
```

```python
x_test=preprocessor.transform(x_test)
```

```python
x_test
```

```python
## apply machine leaarning model training (rando classifier tree)
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score
```

```python
models={
    "Random forest":RandomForestClassifier(),
    "logistic Regression":LogisticRegression(),
    "DecisionTreeClassifier":DecisionTreeClassifier()
}
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train) ## train model

    ## make predictions
    y_train_pred=model.predict(x_train)
    y_test_pred=model.predict(x_test)

    ## training set performance
    model_train_accuracy=accuracy_score(y_train,y_train_pred)
    model_train_f1=f1_score(y_train,y_train_pred)
    model_train_precision=precision_score(y_train,y_train_pred)
    model_train_recall=recall_score(y_train,y_train_pred)
    model_train_roc_auc_score=roc_auc_score(y_train,y_train_pred)

    ## test set performance
    model_test_accuracy=accuracy_score(y_test,y_test_pred)
    model_test_f1=f1_score(y_test,y_test_pred)
    model_test_precision=precision_score(y_test,y_test_pred)
    model_test_recall=recall_score(y_test,y_test_pred)
    model_test_roc_auc_score=roc_auc_score(y_test,y_test_pred)

    ## 
    print(list(models.keys())[i])

    print("model perforance for training set")
    print("- accuracy:{:.4f}".format(model_train_accuracy))
    print("=F1 score: {:.4f}".format(model_train_f1))
    
    print("- precision: {:.4f}".format(model_train_precision))
    print("- Recall : {:.4f}".format(model_train_recall))
    print("- Roc Auc Score: {:.4f}".format(model_train_roc_auc_score))

    print("------------------------------------------")
    print("model perforance for test set")
    print("- accuracy:{:.4f}".format(model_test_accuracy))
    print("=F1 score: {:.4f}".format(model_test_f1))
    
    print("- precision: {:.4f}".format(model_test_precision))
    print("- Recall : {:.4f}".format(model_test_recall))
    print("- Roc Auc Score: {:.4f}".format(model_test_roc_auc_score))

    print("============================================================")
```

```python
## hyper paramter training
rf_params={"max_depth":[5,8,None,10],
          "max_features":[5,7,"auto",8],
          "min_samples_split":[2,8,15,20],
          "n_estimators":[100,200,500,1000]}
```

```python
randomcv_models=[
    ("rf",RandomForestClassifier(),rf_params),
]
```

```python
randomcv_models
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_param = {}

for name, model, params in randomcv_models:
    print(f"================ Tunning {name} ================")
    
    # 💡 Safety Check: Make sure n_iter isn't larger than total possible combinations
    # Optional but highly recommended if your search space is tight
    
    random = RandomizedSearchCV(
        estimator=model,
        param_distributions=params,
        n_iter=20,          # 💡 Reduced from 100 to 20 for faster execution; increase later if needed
        cv=3,
        verbose=2,
        n_jobs=-1,          # Uses all CPU cores
        random_state=42     # 💡 Added for reproducible, consistent results
    )
    
    random.fit(x_train, y_train)
    model_param[name] = random.best_params_

# Corrected printing loop
for model_name in model_param:
    print(f"\n----------------- Best params for {model_name} -----------------")
    print(model_param[model_name])
```
